```python
# import modules
import os
import pandas as pd
```

```python
cityData = pd.read_csv('Resources/output_data/cities.csv')
cityData.head()
```

cityData.rename(index={'0', 'City Id'})
cityData = cityData[['City Id','City','Country','Lat','Lng','Max Temp', 'Humidity','Cloudiness','Wind Speed','Date']]
cityData.head()


```python
cityData.count()
```

```python
cityData.dropna(inplace=True)
cityData.count()
```

```python

cityData['Date'] = pd.to_datetime(cityData['Date'],unit='s')
cityData['Date'] = cityData['Date'].dt.strftime('%d-%m-%Y %H:%M')
cityData.head()
```

```python
cityData.to_html('tableData.html', index=False, classes=['table', 'table-striped', 'table-hover'])
```
